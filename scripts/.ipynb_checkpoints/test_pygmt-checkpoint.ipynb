```python
import pygmt
from pygmt.datasets import load_earth_relief
```

```python
grid = load_earth_relief(resolution="05m", region=[0,360,-90,-60],registration="gridline")
```

```python
fig = pygmt.Figure()
#fig.coast(region="g", frame=True, shorelines=1)
fig.grdimage(grid=grid, projection="S0.0/-90.0/50/5i", cmap="geo", frame=True)
fig.grdcontour(grid=grid, levels=1000, annotation=1000)
fig.coast(shorelines="2p")
fig.colorbar(frame=["a1000", "x+lTopographie", "y+lm"])
# fig.coast(
#     region=[0,360,-90,-60],
#     shorelines=True,
#     land="lightgreen",
#     water="lightblue",
#     projection="S0.0/-90.0/50/5i",
# )
fig.show()
```
