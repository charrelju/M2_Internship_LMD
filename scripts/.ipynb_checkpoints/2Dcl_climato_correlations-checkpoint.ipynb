```python
import netCDF4
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MultipleLocator
```

```python
# OBS
path = '/home/jcharrel/Documents/stage_M2/data/CALIPSO/'
ncfile_obs=path+"CALIPSO_2Dcl_climato_concatenated_averaged_2008_2021_not_2016_r144x90.nc"
nc_obs = netCDF4.Dataset(ncfile_obs)
```

```python
# LMDZ6A
path = '/home/jcharrel/Documents/stage_M2/data/LMDZ6A/'
ncfile_lmdz6a=path+"LmdzOr6Actrl_calipso_r144x90.nc"
nc_lmdz6a = netCDF4.Dataset(ncfile_lmdz6a)
```

```python
# GRID
lon = nc_obs.variables['lon'][:]
lat = nc_obs.variables['lat'][:]

lat_id = (lat <= -60) 
lat = lat[lat_id]

lon_est = lon[lon<=150]
lon_peninsule = lon[(lon> 280) & (lon<=305)]
lon_ouest = lon[(lon>150) & (lon<=170) | (lon>210)&(lon<=280) | (lon>340) &(lon<=360)]
lon_ice_shelves = lon[(lon>170) & (lon <=210) | (lon>305) & (lon <=340)]

lon_est_id = lon<=150
lon_peninsule_id = (lon> 280) & (lon<=305)
lon_ouest_id = (lon>150) & (lon<=170) | (lon>210)&(lon<=280) | (lon>340) &(lon<=360)
lon_ice_shelves_id = (lon>170) & (lon <=210) | (lon>305) & (lon <=340)
```

```python
# VARIABLES - OBS
cllobs = np.mean(nc_obs.variables['cllcalipso'][:,lat_id,:],axis=0)*100
clmobs = np.mean(nc_obs.variables['clmcalipso'][:,lat_id,:],axis=0)*100
clhobs = np.mean(nc_obs.variables['clhcalipso'][:,lat_id,:],axis=0)*100
cltobs = np.mean(nc_obs.variables['cltcalipso'][:,lat_id,:],axis=0)*100

# VARIABLES - LMDZ6A
clllmdz6a = np.mean(nc_lmdz6a.variables['cllcalipso'][:,lat_id,:],axis=0)*100
clmlmdz6a = np.mean(nc_lmdz6a.variables['clmcalipso'][:,lat_id,:],axis=0)*100
clhlmdz6a = np.mean(nc_lmdz6a.variables['clhcalipso'][:,lat_id,:],axis=0)*100
cltlmdz6a = np.mean(nc_lmdz6a.variables['cltcalipso'][:,lat_id,:],axis=0)*100
```

```python
# FIRST DATA PROCESSING
# OBS
cllobs[(cllobs < 1) | (cllobs > 100)] = np.nan
clmobs[(clmobs < 1) | (clmobs > 100)] = np.nan
clhobs[(clhobs < 1) | (clhobs > 100)] = np.nan
cltobs[(cltobs < 1) | (cltobs > 100)] = np.nan

# LMDZ6A
clllmdz6a[(clllmdz6a < 1) | (clllmdz6a > 100)] = np.nan
clmlmdz6a[(clmlmdz6a < 1) | (clmlmdz6a > 100)] = np.nan
clhlmdz6a[(clhlmdz6a < 1) | (clhlmdz6a > 100)] = np.nan
cltlmdz6a[(cltlmdz6a < 1) | (cltlmdz6a > 100)] = np.nan
```

```python
# OBS
# EAST
cllobs_est = cllobs[:,lon_est_id]
clmobs_est = clmobs[:,lon_est_id]
clhobs_est = clhobs[:,lon_est_id]
cltobs_est = cltobs[:,lon_est_id]

# PENINSULA
cllobs_peninsule = cllobs[:,lon_peninsule_id]
clmobs_peninsule = clmobs[:,lon_peninsule_id]
clhobs_peninsule = clhobs[:,lon_peninsule_id]
cltobs_peninsule = cltobs[:,lon_peninsule_id]

# WEST
cllobs_ouest = cllobs[:,lon_ouest_id]
clmobs_ouest = clmobs[:,lon_ouest_id]
clhobs_ouest = clhobs[:,lon_ouest_id]
cltobs_ouest = cltobs[:,lon_ouest_id]

# ICE SHELVES
cllobs_ice_shelves = cllobs[:,lon_ice_shelves_id]
clmobs_ice_shelves = clmobs[:,lon_ice_shelves_id]
clhobs_ice_shelves = clhobs[:,lon_ice_shelves_id]
cltobs_ice_shelves = cltobs[:,lon_ice_shelves_id]

# LMDZ6A
# EAST
clllmdz6a_est = clllmdz6a[:,lon_est_id]
clmlmdz6a_est = clmlmdz6a[:,lon_est_id]
clhlmdz6a_est = clhlmdz6a[:,lon_est_id]
cltlmdz6a_est = cltlmdz6a[:,lon_est_id]

# PENINSULA
clllmdz6a_peninsule = clllmdz6a[:,lon_peninsule_id]
clmlmdz6a_peninsule = clmlmdz6a[:,lon_peninsule_id]
clhlmdz6a_peninsule = clhlmdz6a[:,lon_peninsule_id]
cltlmdz6a_peninsule = cltlmdz6a[:,lon_peninsule_id]

# WEST
clllmdz6a_ouest = clllmdz6a[:,lon_ouest_id]
clmlmdz6a_ouest = clmlmdz6a[:,lon_ouest_id]
clhlmdz6a_ouest = clhlmdz6a[:,lon_ouest_id]
cltlmdz6a_ouest = cltlmdz6a[:,lon_ouest_id]

# ICE SHELVES
clllmdz6a_ice_shelves = clllmdz6a[:,lon_ice_shelves_id]
clmlmdz6a_ice_shelves = clmlmdz6a[:,lon_ice_shelves_id]
clhlmdz6a_ice_shelves = clhlmdz6a[:,lon_ice_shelves_id]
cltlmdz6a_ice_shelves = cltlmdz6a[:,lon_ice_shelves_id]
```

```python
# SECOND DATA PROCESSING --> KEEP ONLY NON NAN VALUES, OBTAIN A 1D LIST

# TOTAL
idxl = np.isfinite(cllobs) & np.isfinite(clllmdz6a)
idxm = np.isfinite(clmobs) & np.isfinite(clmlmdz6a)
idxh = np.isfinite(clhobs) & np.isfinite(clhlmdz6a)
idxt = np.isfinite(cltobs) & np.isfinite(cltlmdz6a)
cllobs = cllobs[idxl]
clmobs = clmobs[idxm]
clhobs = clhobs[idxh]
cltobs = cltobs[idxt]
clllmdz6a  = clllmdz6a[idxl]
clmlmdz6a  = clmlmdz6a[idxm]
clhlmdz6a  = clhlmdz6a[idxh]
cltlmdz6a  = cltlmdz6a[idxt]

# EAST
idxl = np.isfinite(cllobs_est) & np.isfinite(clllmdz6a_est)
idxm = np.isfinite(clmobs_est) & np.isfinite(clmlmdz6a_est)
idxh = np.isfinite(clhobs_est) & np.isfinite(clhlmdz6a_est)
idxt = np.isfinite(cltobs_est) & np.isfinite(cltlmdz6a_est)
cllobs_est = cllobs_est[idxl]
clmobs_est = clmobs_est[idxm]
clhobs_est = clhobs_est[idxh]
cltobs_est = cltobs_est[idxt]
clllmdz6a_est  = clllmdz6a_est[idxl]
clmlmdz6a_est  = clmlmdz6a_est[idxm]
clhlmdz6a_est  = clhlmdz6a_est[idxh]
cltlmdz6a_est  = cltlmdz6a_est[idxt]

# PENINSULE
idxl = np.isfinite(cllobs_peninsule) & np.isfinite(clllmdz6a_peninsule)
idxm = np.isfinite(clmobs_peninsule) & np.isfinite(clmlmdz6a_peninsule)
idxh = np.isfinite(clhobs_peninsule) & np.isfinite(clhlmdz6a_peninsule)
idxt = np.isfinite(cltobs_peninsule) & np.isfinite(cltlmdz6a_peninsule)
cllobs_peninsule = cllobs_peninsule[idxl]
clmobs_peninsule = clmobs_peninsule[idxm]
clhobs_peninsule = clhobs_peninsule[idxh]
cltobs_peninsule = cltobs_peninsule[idxt]
clllmdz6a_peninsule  = clllmdz6a_peninsule[idxl]
clmlmdz6a_peninsule  = clmlmdz6a_peninsule[idxm]
clhlmdz6a_peninsule  = clhlmdz6a_peninsule[idxh]
cltlmdz6a_peninsule  = cltlmdz6a_peninsule[idxt]

# WEST
idxl = np.isfinite(cllobs_ouest) & np.isfinite(clllmdz6a_ouest)
idxm = np.isfinite(clmobs_ouest) & np.isfinite(clmlmdz6a_ouest)
idxh = np.isfinite(clhobs_ouest) & np.isfinite(clhlmdz6a_ouest)
idxt = np.isfinite(cltobs_ouest) & np.isfinite(cltlmdz6a_ouest)
cllobs_ouest = cllobs_ouest[idxl]
clmobs_ouest = clmobs_ouest[idxm]
clhobs_ouest = clhobs_ouest[idxh]
cltobs_ouest = cltobs_ouest[idxt]
clllmdz6a_ouest  = clllmdz6a_ouest[idxl]
clmlmdz6a_ouest  = clmlmdz6a_ouest[idxm]
clhlmdz6a_ouest  = clhlmdz6a_ouest[idxh]
cltlmdz6a_ouest  = cltlmdz6a_ouest[idxt]

# ICE SHELVES
idxl = np.isfinite(cllobs_ice_shelves) & np.isfinite(clllmdz6a_ice_shelves)
idxm = np.isfinite(clmobs_ice_shelves) & np.isfinite(clmlmdz6a_ice_shelves)
idxh = np.isfinite(clhobs_ice_shelves) & np.isfinite(clhlmdz6a_ice_shelves)
idxt = np.isfinite(cltobs_ice_shelves) & np.isfinite(cltlmdz6a_ice_shelves)
cllobs_ice_shelves = cllobs_ice_shelves[idxl]
clmobs_ice_shelves = clmobs_ice_shelves[idxm]
clhobs_ice_shelves = clhobs_ice_shelves[idxh]
cltobs_ice_shelves = cltobs_ice_shelves[idxt]
clllmdz6a_ice_shelves  = clllmdz6a_ice_shelves[idxl]
clmlmdz6a_ice_shelves  = clmlmdz6a_ice_shelves[idxm]
clhlmdz6a_ice_shelves  = clhlmdz6a_ice_shelves[idxh]
cltlmdz6a_ice_shelves  = cltlmdz6a_ice_shelves[idxt]
```

```python
plt.figure(figsize=(12,6))
plt.scatter(cllobs,clllmdz6a, s=5, c='r', marker='.', label='Basse (< 3.2 km)', alpha=0.5)
a,b = np.polyfit(cllobs, clllmdz6a, 1)
r = np.corrcoef(cllobs,clllmdz6a)[0,1]
plt.plot(cllobs, a*cllobs+b, '-', c='r')
r2 =r*r
print('Couverture basse : a={}, b={}, r={}, r2={}'.format(np.round(a,2),np.round(b,2),np.round(r,2), np.round(r2,2)))

plt.scatter(clmobs,clmlmdz6a, s=5, c='g', marker='.', label='Moyenne (3.2 - 6.5 km)', alpha=0.5)
a,b = np.polyfit(clmobs, clmlmdz6a, 1)
r = np.corrcoef(clmobs,clmlmdz6a)[0,1]
plt.plot(clmobs, a*clmobs+b, '-', c='g')
r2 =r*r
print('Couverture moyenne : a={}, b={}, r={}, r2={}'.format(np.round(a,2),np.round(b,2),np.round(r,2), np.round(r2,2)))

plt.scatter(clhobs,clhlmdz6a, s=5, c='b', marker='.', label='Haute (> 6.5 km)', alpha=0.5)
a,b = np.polyfit(clhobs, clhlmdz6a, 1)
r = np.corrcoef(clhobs,clhlmdz6a)[0,1]
plt.plot(clhobs, a*clhobs+b, '-', c='b')
r2 =r*r
print('Couverture haute : a={}, b={}, r={}, r2={}'.format(np.round(a,2),np.round(b,2),np.round(r,2), np.round(r2,2)))

plt.scatter(cltobs,cltlmdz6a, s=5, c='k', marker='.', label='Totale', alpha=0.5)
a,b = np.polyfit(cltobs, cltlmdz6a, 1)
r = np.corrcoef(cltobs,cltlmdz6a)[0,1]
plt.plot(cltobs, a*cltobs+b, '-', c='k')
r2 =r*r
print('Couverture totale : a={}, b={}, r={}, r2={}'.format(np.round(a,2),np.round(b,2),np.round(r,2), np.round(r2,2)))

plt.legend()
plt.xlim(0,100)
plt.ylim(0,100)
plt.axline((0, 0), slope=1, c='grey', linestyle='--', label='x=y')
plt.xlabel('Couverture nuageuse CALIPSO-GOCCP', size=15)
plt.ylabel('Couverture nuageuse LMDZ6A', size=15)
plt.tick_params(axis='both', labelsize=12)
plt.tight_layout()
path_fig = '/home/jcharrel/Documents/stage_M2/figures/CALIPSO_climato/'
plt.savefig(path_fig+'2Dcl_climato_correlation_total.png',dpi=300)
```

```python
fig, axes = plt.subplots(figsize=(16,9), nrows=3, ncols=5)
pltcounter = 0 ; altcounter = 0 ; counter=0

fig.text(0.04, 0.825, 'Haute (> 6.5 km)', c='b', fontsize=16, va='center', ha='center', rotation='vertical')
fig.text(0.04, 0.5, 'Moyenne (3.2 - 6.5 km)', c='g', fontsize=16, va='center', ha='center', rotation='vertical')
fig.text(0.04, 0.2, 'Basse (< 3.2 km)', c='r', fontsize=16, va='center', ha='center', rotation='vertical')
label=None

for plotid in axes.flat:
    pltcounter = pltcounter + 1
    altcounter = (pltcounter - 1) % 4 + 1
        
    if pltcounter == 1 :
        plotid.set_title('total', size=15)
        clobs_plot = clhobs
        cllmdz6a_plot = clhlmdz6a
    if pltcounter == 2 : 
        plotid.set_title('est', size=15)
        clobs_plot = clhobs_est
        cllmdz6a_plot = clhlmdz6a_est
    if pltcounter == 3 : 
        plotid.set_title('péninsule', size=15)
        clobs_plot = clhobs_peninsule
        cllmdz6a_plot = clhlmdz6a_peninsule
    if pltcounter == 4 : 
        plotid.set_title('ouest', size=15)
        clobs_plot = clhobs_ouest
        cllmdz6a_plot = clhlmdz6a_ouest
    if pltcounter == 5 :
        plotid.set_title('plateformes de glace', size=15)
        clobs_plot = clhobs_ice_shelves
        cllmdz6a_plot = clhlmdz6a_ice_shelves
    if pltcounter == 6 :
        clobs_plot = clmobs
        cllmdz6a_plot = clmlmdz6a
    if pltcounter == 7 :
        clobs_plot = clmobs_est
        cllmdz6a_plot = clmlmdz6a_est
    if pltcounter == 8 :
        clobs_plot = clmobs_peninsule
        cllmdz6a_plot = clmlmdz6a_peninsule
    if pltcounter == 9 :
        clobs_plot = clmobs_ouest
        cllmdz6a_plot = clmlmdz6a_ouest
    if pltcounter == 10 :
        clobs_plot = clmobs_ice_shelves
        cllmdz6a_plot = clmlmdz6a_ice_shelves
    if pltcounter == 11 :
        clobs_plot = cllobs
        cllmdz6a_plot = clllmdz6a
    if pltcounter == 12 :
        clobs_plot = cllobs_est
        cllmdz6a_plot = clllmdz6a_est
    if pltcounter == 13 :
        clobs_plot = cllobs_peninsule
        cllmdz6a_plot = clllmdz6a_peninsule
    if pltcounter == 14 :
        clobs_plot = cllobs_ouest
        cllmdz6a_plot = clllmdz6a_ouest
    if pltcounter == 15 :
        clobs_plot = cllobs_ice_shelves
        cllmdz6a_plot = clllmdz6a_ice_shelves
        label = 'x=y'
    if pltcounter in [11,12,13,14,15] :
        plotid.set_xlabel('CN CALIPSO-GOCCP (%)', size=12)
    if pltcounter in [1,6,11]: 
        plotid.set_ylabel('CN LMDZ6A (%)', size=12)

    if pltcounter in [1,2,3,4,5]:
        color = 'b'
        plotid.set_xticks(np.arange(0,51,10))
        plotid.set_xlim(0,50)
        plotid.set_ylim(0,50)
        plotid.xaxis.set_minor_locator(MultipleLocator(5))
        plotid.yaxis.set_minor_locator(MultipleLocator(5))
    if pltcounter in [6,7,8,9,10]:
        color = 'g'
        plotid.set_xlim(0,45)
        plotid.set_ylim(0,45)
        plotid.xaxis.set_minor_locator(MultipleLocator(5))
        plotid.yaxis.set_minor_locator(MultipleLocator(5))
    if pltcounter in [11,12,13,14,15]:
        color = 'r'
        # plotid.set_xticks(np.arange(0,101,20))
        plotid.set_xlim(0,80)
        plotid.set_ylim(0,80)
        plotid.xaxis.set_minor_locator(MultipleLocator(10))
        plotid.yaxis.set_minor_locator(MultipleLocator(10))

    plotid.scatter(clobs_plot,cllmdz6a_plot, s=2, c=color, marker='.', alpha=0.5)
    a,b = np.polyfit(clobs_plot,cllmdz6a_plot,1)
    r = np.corrcoef(clobs_plot, cllmdz6a_plot)[0, 1]
    r2 = r*r
    plotid.plot(clobs_plot, a*clobs_plot+b, c='k')
    plotid.axline((0, 0), slope=1, c='grey', linestyle='--', label=label)

    plotid.tick_params(axis='both', labelsize=12)
    plotid.text(0.05,0.8,'n={}\nr={}\nr2={}'.format(len(clobs_plot),np.round(r,2),np.round(r2,2)), transform = plotid.transAxes)
    #plotid.legend(loc='upper left')

    # plotid.xaxis.set_major_locator(MultipleLocator(10))
    # plotid.xaxis.set_minor_locator(MultipleLocator(5))
    # plotid.yaxis.set_major_locator(MultipleLocator(5))
    # plotid.yaxis.set_minor_locator(MultipleLocator(1))

#fig.suptitle('GOCCP (2007-2021, not 2016) & LMDZ6A (1980-1999) climatology', fontsize=20)
fig.tight_layout()
fig.subplots_adjust(right=0.89, left=0.1, wspace=0.3)
#fig.legend()
path_fig = '/home/jcharrel/Documents/stage_M2/figures/CALIPSO_climato/'
fig.savefig(path_fig+'2Dcl_climato_correlation_zones.png',dpi=300)
```
