```python
import netCDF4
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import cartopy.crs as ccrs
from cartopy.util import add_cyclic_point
```

```python
# OBS
path = '/home/jcharrel/Documents/stage_M2/data/CALIPSO/'
ncfile_cl=path+"CALIPSO_2Dcl_seasonal_averaged_concatenated_2008_2022_r144x90.nc"
nc_cl = netCDF4.Dataset(ncfile_cl)
```

```python
# MASK
path = '/home/jcharrel/Documents/stage_M2/data/LMDZ6A/'
ncfile_mask=path+"LmdzOr6Actrl_SE_1980_1999_1M_histmth_phis_r144x90.nc"
nc_mask = netCDF4.Dataset(ncfile_mask)
```

```python
# GRID
lon = nc_cl.variables['lon'][:]
lat = nc_cl.variables['lat'][:]

lat_austral = lat[lat <= -30]
lat_antarctica = lat[lat <= -60]

lat_id_austral = lat <= -30
lat_id_antarctica = lat <= -60

import matplotlib.path as mpath
theta = np.linspace(0, 2*np.pi, 100)
map_circle = mpath.Path(np.vstack([np.sin(theta), np.cos(theta)]).T * 0.5 + [0.5, 0.5])
```

```python
# VARIABLES

# CALIPSO low
clmobs_austral = nc_cl.variables['clmcalipso'][:,lat_id_austral,:]
clmobs_antarctica = nc_cl.variables['clmcalipso'][:,lat_id_antarctica,:]

# GEOPOTENTIAL
g = 9.81 # m/s
alt_austral = nc_mask.variables['phis'][:,lat_id_austral,:]/g #m
alt_antarctica = nc_mask.variables['phis'][:,lat_id_antarctica,:]/g #m
```

```python
# MASK

mask_austral = np.where(alt_austral > 0.1, 1, np.nan) #pour garder que la partie oc√©an austral
mask_antarctica = np.where(alt_antarctica <  0.1, 1, np.nan) #pour garder que la partie antarctique
```

```python
clmobs_austral = np.ma.masked_where(mask_austral >= 0, clmobs_austral)
clmobs_antarctica = np.ma.masked_where(mask_antarctica >=0, clmobs_antarctica)
```

```python
# PREUVE QUE CA MARCHE 
fig, axes = plt.subplots(figsize=(16,9), nrows=1, ncols=1, subplot_kw={'projection': ccrs.SouthPolarStereo(central_longitude=0.)})
plt.contourf(lon,lat_austral,clmobs_austral[0,:,:], transform=ccrs.PlateCarree(), cmap='jet')

fig, axes = plt.subplots(figsize=(16,9), nrows=1, ncols=1, subplot_kw={'projection': ccrs.SouthPolarStereo(central_longitude=0.)})
plt.contourf(lon,lat_antarctica,clmobs_antarctica[0,:,:], transform=ccrs.PlateCarree(), cmap='jet')
```
