```python
import netCDF4
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import cartopy.crs as ccrs
from cartopy.util import add_cyclic_point
```

```python
# OBS
path = '/home/jcharrel/Documents/stage_M2/data/CALIPSO/'
ncfile_obs=path+"CALIPSO_2Dcl_climato_concatenated_averaged_2008_2021_not_2016_r144x90.nc"
nc_obs = netCDF4.Dataset(ncfile_obs)
```

```python
# MASK
path = '/home/jcharrel/Documents/stage_M2/data/LMDZ6A/'
ncfile_mask=path+"LmdzOr6Actrl_SE_1980_1999_1M_histmth_phis_r144x90.nc"
nc_mask = netCDF4.Dataset(ncfile_mask)
```

```python
# LMDZ6A
path = '/home/jcharrel/Documents/stage_M2/data/LMDZ6A/'
ncfile_lmdz6a=path+'LmdzOr6Actrl_calipso_r144x90.nc'
nc_lmdz6a = netCDF4.Dataset(ncfile_lmdz6a)
```

```python
# GRID
lon = nc_obs.variables['lon'][:]
lat = nc_obs.variables['lat'][:]

lat_id = (lat <= -60) 
lat = lat[lat_id]

import matplotlib.path as mpath
theta = np.linspace(0, 2*np.pi, 100)
map_circle = mpath.Path(np.vstack([np.sin(theta), np.cos(theta)]).T * 0.5 + [0.5, 0.5])
```

```python
# VARIABLES - OBS
clcobs = nc_obs.variables['clccalipso'][0,lat_id,:]
cllobs = nc_obs.variables['cllcalipso'][0,lat_id,:]
clmobs = nc_obs.variables['clmcalipso'][0,lat_id,:]
clhobs = nc_obs.variables['clhcalipso'][0,lat_id,:]
cltobs = nc_obs.variables['cltcalipso'][0,lat_id,:]
allresults = np.dstack((cllobs,clmobs,clhobs,clcobs,cltobs))
obs = np.dstack((cllobs,clmobs,clhobs,cltobs))

# VARIABLES - LMDZ6A
clllmdz6a = nc_lmdz6a.variables['cllcalipso'][0,lat_id,:]
clllmdz6a[clllmdz6a>1e5] = np.nan
clmlmdz6a = nc_lmdz6a.variables['clmcalipso'][0,lat_id,:]
clmlmdz6a[clmlmdz6a>1e5] = np.nan
clhlmdz6a = nc_lmdz6a.variables['clhcalipso'][0,lat_id,:]
clhlmdz6a[clhlmdz6a>1e5] = np.nan
cltlmdz6a = nc_lmdz6a.variables['cltcalipso'][0,lat_id,:]
cltlmdz6a[cltlmdz6a>1e5] = np.nan
lmdz6a = np.dstack((clllmdz6a,clmlmdz6a,clhlmdz6a,cltlmdz6a))

# DIFFERENCE MODEL-OBS
clldiff = clllmdz6a-cllobs
clmdiff = clmlmdz6a-clmobs
clhdiff = clhlmdz6a-clhobs
cltdiff = cltlmdz6a-cltobs
diff = np.ma.dstack((clldiff, clmdiff, clhdiff, cltdiff))

total = np.ma.dstack((obs,lmdz6a,diff*100))

# ALTITUDE
g = 9.81 # m/s
alt = nc_mask.variables['phis'][0,lat_id,:]/g #m
```

```python
fig, axes = plt.subplots(figsize=(16,9), nrows=3, ncols=4, subplot_kw={'projection': ccrs.SouthPolarStereo(central_longitude=0)})
pltcounter = 0

fig.text(0.04, 0.82, 'OBS', fontsize=16, va='center', ha='center', rotation='vertical')
fig.text(0.04, 0.5, 'LMDZ6A', fontsize=16, va='center', ha='center', rotation='vertical')
fig.text(0.03, 0.17, 'DIFFÉRENCE \n (LMDZ6A - OBS)', fontsize=16, va='center', ha='center', rotation='vertical')

for plotid in axes.flat:
    pltcounter = pltcounter + 1
    if pltcounter == 1:
        plotid.set_title("a) Basse (< 3.2 km)", fontsize=15)
    elif pltcounter == 2:
        plotid.set_title("b) Moyenne (3.2 - 6.5 km)", fontsize=15)
    elif pltcounter == 3:
        plotid.set_title("c) Haute (> 6.5 km)", fontsize=15)
    elif pltcounter == 4:
        plotid.set_title("d) Totale", fontsize=15)

    clplot = total[:,:,pltcounter-1]
    clplt, lonplt = add_cyclic_point(clplot, coord=lon) # filling the gap at longitude 0

    plotid.set_extent([-180, 180, np.min(lat), np.max(lat)], ccrs.PlateCarree())
    plotid.set_boundary(map_circle, transform=plotid.transAxes)
    
    if pltcounter <= 8 :
        clevs = np.array([0,1,2,4,6,8,10,15,20,30,40,50,60,70,80,90,100])
        norm = mpl.colors.BoundaryNorm(boundaries=clevs, ncolors=256)
        cmap = 'jet'
        cs1 = plotid.contourf(lonplt, lat, clplt*100, clevs, transform=ccrs.PlateCarree(),  
                         norm = norm, 
                         cmap=cmap)
    else :
        clevs = np.array([-0.4, -0.35, -0.3, -0.25, -0.2, -0.15, -0.1, -0.05, 0, 0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4])*100
        cmap='seismic'
        cs2 = plotid.contourf(lonplt, lat, clplt,  clevs, transform=ccrs.PlateCarree(),  
                         norm = None, 
                         cmap=cmap)
    
    plotid.coastlines(resolution="110m",linewidth=1)
    alt_contours = plotid.contour(lon, lat, alt, colors='black', linestyles=':', levels=[1000,2000,3000,4000], linewidths=0.5,  transform=ccrs.PlateCarree())

    # Ajouter les lignes de grille
    # from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
    # gl = plotid.gridlines(color='grey', linestyle='--', draw_labels=True)
    # gl.xlocator = mpl.ticker.FixedLocator(np.arange(-180, 181, 30))
    # gl.ylocator = mpl.ticker.FixedLocator(np.arange(-90, 91, 40))
    # gl.xformatter = LONGITUDE_FORMATTER
    # gl.yformatter = LATITUDE_FORMATTER
    # gl.xlabel_style = {'rotation': 0}
    # gl.ylabel_style = {'color': 'gray'}

    # Ajouter les labels des altitudes
    #plotid.clabel(alt_contours, inline=True, fontsize=8, fmt='%1.0f m')
    
    # Dumont d'Urville
    #plotid.text(140, -65, 'DDU', transform=ccrs.PlateCarree(), fontsize=9, color='black')
    #plotid.scatter(140, -66, marker='.', s=20, color='k', transform=ccrs.PlateCarree())
    # Concordia
    #plotid.text(125, -74, 'DC', transform=ccrs.PlateCarree(), fontsize=9, color='black')
    #plotid.scatter(123, -75, marker='.', s=20, color='k', transform=ccrs.PlateCarree())

#fig.suptitle('GOCCP (2007-2021, not 2016) & LMDZ6A (1980-1999) climatology', fontsize=20)

fig.tight_layout()
fig.subplots_adjust(right=0.8, left=0.05, wspace=0.1)

line = plt.Line2D([0.05, 0.8], [0.98/3, 0.98/3], color='black', linestyle='--', linewidth=2, transform=fig.transFigure)
fig.add_artist(line)

cbar_ax1 = fig.add_axes([0.84, 0.35, 0.008, 0.6])
cbar_ax2 = fig.add_axes([0.84, 0.01, 0.008, 0.32])
cb1 = fig.colorbar(cs1, cax=cbar_ax1, ticks=np.array([0,1,2,4,6,8,10,15,20,30,40,50,60,70,80,90,100]))
cb2 = fig.colorbar(cs2, cax=cbar_ax2, ticks=np.array([-0.4, -0.35, -0.3, -0.25, -0.2, -0.15, -0.1, -0.05, 0, 0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4])*100)
cb1.ax.tick_params(labelsize=12)
cb2.ax.tick_params(labelsize=12)
cb1.set_label('Couverture nuageuse (%)', fontsize=14)
cb2.set_label('Différence (%)', fontsize=14)

path_fig = '/home/jcharrel/Documents/stage_M2/figures/CALIPSO_climato/'
plt.savefig(path_fig+'cl_climato_antarctica_obs_lmdz6a.png',dpi=300)
```
