```python
import netCDF4
import matplotlib.pyplot as plt
import numpy as np
```

```python
# OBS
path = '/home/jcharrel/Documents/stage_M2/data/CALIPSO/'
ncfile_obs=path+"CALIPSO_2Dcl_seasonal_averaged_concatenated_2008_2022_r144x90.nc"
nc_obs = netCDF4.Dataset(ncfile_obs)
```

```python
# MASK
path = '/home/jcharrel/Documents/stage_M2/data/LMDZ6A/'
ncfile_mask=path+"LmdzOr6Actrl_SE_1980_1999_1M_histmth_phis_r144x90.nc"
nc_mask = netCDF4.Dataset(ncfile_mask)
```

```python
# LMDZ6A
path = '/home/jcharrel/Documents/stage_M2/data/LMDZ6A/'
ncfile_lmdz6a=path+'LmdzOr6Actrl_calipso_r144x90.nc'
nc_lmdz6a = netCDF4.Dataset(ncfile_lmdz6a)
```

```python
# GRID
lon = nc_obs.variables['lon'][:]
lat = nc_obs.variables['lat'][:]

lat_id = lat <= -30
lat = lat[lat <= -30]

lon[lon > 180] -= 360
lon_id = np.argsort(lon)
lon = lon[lon_id]

import matplotlib.path as mpath
theta = np.linspace(0, 2*np.pi, 100)
map_circle = mpath.Path(np.vstack([np.sin(theta), np.cos(theta)]).T * 0.5 + [0.5, 0.5])
```

```python
# VARIABLES - OBS
cllobs = nc_obs.variables['cllcalipso'][:,lat_id,lon_id]
clmobs = nc_obs.variables['clmcalipso'][:,lat_id,lon_id]
clhobs = nc_obs.variables['clhcalipso'][:,lat_id,lon_id]
cltobs = nc_obs.variables['cltcalipso'][:,lat_id,lon_id]

# VARIABLE - LMDZ6A
clllmdz6a = nc_lmdz6a.variables['cllcalipso'][:,lat_id,lon_id]
clmlmdz6a = nc_lmdz6a.variables['clmcalipso'][:,lat_id,lon_id]
clhlmdz6a = nc_lmdz6a.variables['clhcalipso'][:,lat_id,lon_id]
cltlmdz6a = nc_lmdz6a.variables['cltcalipso'][:,lat_id,lon_id]
```

```python
# MASK LMDZ6A WHERE THERE IS NO OBS DATA
mask_lat_obs_cll = np.where(cllobs.mask == True, 1, np.nan)
mask_lat_obs_clm = np.where(clmobs.mask == True, 1, np.nan)
mask_lat_obs_clh = np.where(clhobs.mask == True, 1, np.nan)
mask_lat_obs_clt = np.where(cltobs.mask == True, 1, np.nan)

clllmdz6a = np.ma.masked_where(mask_lat_obs_cll >=0, clllmdz6a)
clmlmdz6a = np.ma.masked_where(mask_lat_obs_clm >=0, clmlmdz6a)
clhlmdz6a = np.ma.masked_where(mask_lat_obs_clh >=0, clhlmdz6a)
cltlmdz6a = np.ma.masked_where(mask_lat_obs_clt >=0, cltlmdz6a)
```

```python
# SELECT SEASONS AND MEAN 

# OBS
cllobs_DJF = np.mean(np.mean(cllobs, axis=2)[[0,1,11],:], axis=0)
cllobs_MAM = np.mean(np.mean(cllobs, axis=2)[[2,3,4],:], axis=0)
cllobs_JJA = np.mean(np.mean(cllobs, axis=2)[[5,6,7],:], axis=0)
cllobs_SON = np.mean(np.mean(cllobs, axis=2)[[8,9,10],:], axis=0)
cllobs_seasons = np.ma.dstack((cllobs_DJF, cllobs_MAM, cllobs_JJA, cllobs_SON))
cllobs_mean = np.mean(cllobs_seasons, axis=(0,2))

clmobs_DJF = np.mean(np.mean(clmobs, axis=2)[[0,1,11],:], axis=0)
clmobs_MAM = np.mean(np.mean(clmobs, axis=2)[[2,3,4],:], axis=0)
clmobs_JJA = np.mean(np.mean(clmobs, axis=2)[[5,6,7],:], axis=0)
clmobs_SON = np.mean(np.mean(clmobs, axis=2)[[8,9,10],:], axis=0)
clmobs_seasons = np.ma.dstack((clmobs_DJF, clmobs_MAM, clmobs_JJA, clmobs_SON))
clmobs_mean = np.mean(clmobs_seasons, axis=(0,2))

clhobs_DJF = np.mean(np.mean(clhobs, axis=2)[[0,1,11],:], axis=0)
clhobs_MAM = np.mean(np.mean(clhobs, axis=2)[[2,3,4],:], axis=0)
clhobs_JJA = np.mean(np.mean(clhobs, axis=2)[[5,6,7],:], axis=0)
clhobs_SON = np.mean(np.mean(clhobs, axis=2)[[8,9,10],:], axis=0)
clhobs_seasons = np.ma.dstack((clhobs_DJF, clhobs_MAM, clhobs_JJA, clhobs_SON))
clhobs_mean = np.mean(clhobs_seasons, axis=(0,2))

cltobs_DJF = np.mean(np.mean(cltobs, axis=2)[[0,1,11],:], axis=0)
cltobs_MAM = np.mean(np.mean(cltobs, axis=2)[[2,3,4],:], axis=0)
cltobs_JJA = np.mean(np.mean(cltobs, axis=2)[[5,6,7],:], axis=0)
cltobs_SON = np.mean(np.mean(cltobs, axis=2)[[8,9,10],:], axis=0)
cltobs_seasons = np.ma.dstack((cltobs_DJF, cltobs_MAM, cltobs_JJA, cltobs_SON))
cltobs_mean = np.mean(cltobs_seasons, axis=(0,2))

# LMDZ6A
clllmdz6a_DJF = np.mean(np.mean(clllmdz6a, axis=2)[[0,1,11],:], axis=0)
clllmdz6a_MAM = np.mean(np.mean(clllmdz6a, axis=2)[[2,3,4],:], axis=0)
clllmdz6a_JJA = np.mean(np.mean(clllmdz6a, axis=2)[[5,6,7],:], axis=0)
clllmdz6a_SON = np.mean(np.mean(clllmdz6a, axis=2)[[8,9,10],:], axis=0)
clllmdz6a_seasons = np.ma.dstack((clllmdz6a_DJF, clllmdz6a_MAM, clllmdz6a_JJA, clllmdz6a_SON))
clllmdz6a_mean = np.mean(clllmdz6a_seasons, axis=(0,2))

clmlmdz6a_DJF = np.mean(np.mean(clmlmdz6a, axis=2)[[0,1,11],:], axis=0)
clmlmdz6a_MAM = np.mean(np.mean(clmlmdz6a, axis=2)[[2,3,4],:], axis=0)
clmlmdz6a_JJA = np.mean(np.mean(clmlmdz6a, axis=2)[[5,6,7],:], axis=0)
clmlmdz6a_SON = np.mean(np.mean(clmlmdz6a, axis=2)[[8,9,10],:], axis=0)
clmlmdz6a_seasons = np.ma.dstack((clmlmdz6a_DJF, clmlmdz6a_MAM, clmlmdz6a_JJA, clmlmdz6a_SON))
clmlmdz6a_mean = np.mean(clmlmdz6a_seasons, axis=(0,2))

clhlmdz6a_DJF = np.mean(np.mean(clhlmdz6a, axis=2)[[0,1,11],:], axis=0)
clhlmdz6a_MAM = np.mean(np.mean(clhlmdz6a, axis=2)[[2,3,4],:], axis=0)
clhlmdz6a_JJA = np.mean(np.mean(clhlmdz6a, axis=2)[[5,6,7],:], axis=0)
clhlmdz6a_SON = np.mean(np.mean(clhlmdz6a, axis=2)[[8,9,10],:], axis=0)
clhlmdz6a_seasons = np.ma.dstack((clhlmdz6a_DJF, clhlmdz6a_MAM, clhlmdz6a_JJA, clhlmdz6a_SON))
clhlmdz6a_mean = np.mean(clhlmdz6a_seasons, axis=(0,2))

cltlmdz6a_DJF = np.mean(np.mean(cltlmdz6a, axis=2)[[0,1,11],:], axis=0)
cltlmdz6a_MAM = np.mean(np.mean(cltlmdz6a, axis=2)[[2,3,4],:], axis=0)
cltlmdz6a_JJA = np.mean(np.mean(cltlmdz6a, axis=2)[[5,6,7],:], axis=0)
cltlmdz6a_SON = np.mean(np.mean(cltlmdz6a, axis=2)[[8,9,10],:], axis=0)
cltlmdz6a_seasons = np.ma.dstack((cltlmdz6a_DJF, cltlmdz6a_MAM, cltlmdz6a_JJA, cltlmdz6a_SON))
cltlmdz6a_mean = np.mean(cltlmdz6a_seasons, axis=(0,2))
```

```python
fig, axes = plt.subplots(figsize=(16,9), nrows=2, ncols=2)
pltcounter = 0
for plotid in axes.flat:
    pltcounter = pltcounter + 1
    if pltcounter == 1:
        plotid.set_title("a) Low", size=15)
        plotid.set_ylabel('Cloud cover')
        plotid.plot(lat,cllobs_seasons[0,:,0], c='r')
        plotid.plot(lat,cllobs_seasons[0,:,1], c='orange')
        plotid.plot(lat,cllobs_seasons[0,:,2], c='b')
        plotid.plot(lat,cllobs_seasons[0,:,3], c='g')
        plotid.plot(lat, cllobs_mean, c='k')
        plotid.plot(lat,clllmdz6a_seasons[0,:,0], ':', c='r')
        plotid.plot(lat,clllmdz6a_seasons[0,:,1], ':', c='orange')
        plotid.plot(lat,clllmdz6a_seasons[0,:,2], ':', c='b')
        plotid.plot(lat,clllmdz6a_seasons[0,:,3], ':', c='g')
        plotid.plot(lat, clllmdz6a_mean, ':', c='k')
        plotid.set_xticks(np.arange(-85,-20,5))
        plotid.set_ylim(0,0.8)
        
    elif pltcounter == 2:
        plotid.set_title("b) Mid", size=15)
        plotid.plot(lat,clmobs_seasons[0,:,0], c='r')
        plotid.plot(lat,clmobs_seasons[0,:,1], c='orange')
        plotid.plot(lat,clmobs_seasons[0,:,2], c='b')
        plotid.plot(lat,clmobs_seasons[0,:,3], c='g')
        plotid.plot(lat, clmobs_mean, c='k')
        plotid.plot(lat,clmlmdz6a_seasons[0,:,0], ':', c='r')
        plotid.plot(lat,clmlmdz6a_seasons[0,:,1], ':', c='orange')
        plotid.plot(lat,clmlmdz6a_seasons[0,:,2], ':', c='b')
        plotid.plot(lat,clmlmdz6a_seasons[0,:,3], ':', c='g')
        plotid.plot(lat, clmlmdz6a_mean, ':', c='k')
        plotid.set_xticks(np.arange(-85,-20,5))
        plotid.set_ylim(0,0.45)
        
    elif pltcounter == 3:
        plotid.set_title("c) High", size=15)
        plotid.set_ylabel('Cloud cover')
        plotid.plot(lat,clhobs_seasons[0,:,0], c='r')
        plotid.plot(lat,clhobs_seasons[0,:,1], c='orange')
        plotid.plot(lat,clhobs_seasons[0,:,2], c='b')
        plotid.plot(lat,clhobs_seasons[0,:,3], c='g')
        plotid.plot(lat, clhobs_mean, c='k')
        plotid.plot(lat,clhlmdz6a_seasons[0,:,0], ':', c='r')
        plotid.plot(lat,clhlmdz6a_seasons[0,:,1], ':', c='orange')
        plotid.plot(lat,clhlmdz6a_seasons[0,:,2], ':', c='b')
        plotid.plot(lat,clhlmdz6a_seasons[0,:,3], ':', c='g')
        plotid.plot(lat, clhlmdz6a_mean, ':', c='k')
        plotid.set_xlabel('Latitude')
        plotid.set_xticks(np.arange(-85,-20,5))
        plotid.set_ylim(0,0.5)
        
    elif pltcounter == 4:
        plotid.set_title("d) Total", size=15)
        plotid.plot(lat,cltobs_seasons[0,:,0], c='r', label='OBS DJF')
        plotid.plot(lat,cltobs_seasons[0,:,1], c='orange', label='OBS MAM')
        plotid.plot(lat,cltobs_seasons[0,:,2], c='b', label='OBS JJA')
        plotid.plot(lat,cltobs_seasons[0,:,3], c='g', label='OBS SON')
        plotid.plot(lat, cltobs_mean, c='k', label='OBS annually')
        plotid.plot(lat,cltlmdz6a_seasons[0,:,0], ':', c='r', label='LMDZ6A DJF')
        plotid.plot(lat,cltlmdz6a_seasons[0,:,1], ':', c='orange', label='LMDZ6A MAM')
        plotid.plot(lat,cltlmdz6a_seasons[0,:,2], ':', c='b', label='LMDZ6A JJA')
        plotid.plot(lat,cltlmdz6a_seasons[0,:,3], ':', c='g', label='LMDZ6A SON')
        plotid.plot(lat, cltlmdz6a_mean, ':', c='k', label='LMDZ6A annually')
        plotid.set_xlabel('Latitude')
        plotid.set_xticks(np.arange(-85,-20,5))
        plotid.set_ylim(0.2,1)
        plotid.legend(loc=(0.5,0.1))

fig.suptitle('Zonally averaged GOCCP and LMDZ6A cloud cover', fontsize=20)
fig.tight_layout()
path_fig = '/home/jcharrel/Documents/stage_M2/figures/CALIPSO_seasonal/'
plt.savefig(path_fig+'cl_seasonal_zonal_obs_lmdz6a.png',dpi=300)
```
