```python
import netCDF4
import matplotlib.pyplot as plt
import numpy as np
```

```python
# OBS
path = '/home/jcharrel/Documents/stage_M2/data/CALIPSO/'
ncfile_obs=path+"CALIPSO_2Dcl_seasonal_averaged_concatenated_2006_2022_r144x90.nc"
nc_obs = netCDF4.Dataset(ncfile_obs)
```

```python
# MASK
path = '/home/jcharrel/Documents/stage_M2/data/LMDZ6A/'
ncfile_mask=path+"LmdzOr6Actrl_SE_1980_1999_1M_histmth_phis_r144x90.nc"
nc_mask = netCDF4.Dataset(ncfile_mask)
```

```python
# LMDZ6A FOR T500, SLP, Z500, PRW, PRLW, PRSW
path = '/home/jcharrel/Documents/stage_M2/data/LMDZ6A/'
ncfile_lmdz6a=path+'LmdzOr6Actrl_SE_1980_1999_1M_histmth_slp_t500_z500_prw_prlw_prsw_r144x90.nc'
nc_lmdz6a = netCDF4.Dataset(ncfile_lmdz6a)
```

```python
# GRID
lon = nc_obs.variables['lon'][:]
lat = nc_obs.variables['lat'][:]

lat_austral = lat[lat <= -30]
lat_antarctica = lat[lat <= -60]

lat_id_austral = lat <= -30
lat_id_antarctica = lat <= -60

import matplotlib.path as mpath
theta = np.linspace(0, 2*np.pi, 100)
map_circle = mpath.Path(np.vstack([np.sin(theta), np.cos(theta)]).T * 0.5 + [0.5, 0.5])
```

```python
# GEOPOTENTIAL FOR MASK
g = 9.81 # m/s
alt_austral = nc_mask.variables['phis'][:,lat_id_austral,:]/g #m
alt_antarctica = nc_mask.variables['phis'][:,lat_id_antarctica,:]/g #m
alt_above_3000m = nc_mask.variables['phis'][:,lat_id_antarctica,:]/g #m
```

```python
# MASK
mask_antarctica = np.where(alt_austral > 0.1, 1, np.nan) #pour masquer la partie antarctique
mask_austral = np.where(alt_antarctica <  0.1, 1, np.nan) #pour masquer la partie ocÃ©an austral
mask_above_3000m_for_mid_cloud = np.where(alt_above_3000m > 3000, 1, np.nan)
```

```python
# VARIABLE - LMDZ6A
# AUSTRAL
prwlmdz6a_austral = nc_lmdz6a.variables['prw'][:,lat_id_austral,:]
prlwlmdz6a_austral = nc_lmdz6a.variables['prlw'][:,lat_id_austral,:]
prswlmdz6a_austral = nc_lmdz6a.variables['prsw'][:,lat_id_austral,:]
# ANTARCTICA
prwlmdz6a_antarctica = nc_lmdz6a.variables['prw'][:,lat_id_antarctica,:]
prlwlmdz6a_antarctica = nc_lmdz6a.variables['prlw'][:,lat_id_antarctica,:]
prswlmdz6a_antarctica = nc_lmdz6a.variables['prsw'][:,lat_id_antarctica,:]
```

```python
# MASK LMDZ6A
# AUSTRAL
prwlmdz6a_austral = np.ma.masked_where(mask_antarctica >=0, prwlmdz6a_austral)
prlwlmdz6a_austral = np.ma.masked_where(mask_antarctica >=0, prlwlmdz6a_austral)
prswlmdz6a_austral = np.ma.masked_where(mask_antarctica >=0, prswlmdz6a_austral)
# ANTARCTICA
prwlmdz6a_antarctica = np.ma.masked_where(mask_austral >=0, prwlmdz6a_antarctica)
prlwlmdz6a_antarctica = np.ma.masked_where(mask_austral >=0, prlwlmdz6a_antarctica)
prswlmdz6a_antarctica = np.ma.masked_where(mask_austral >=0, prswlmdz6a_antarctica)
```

```python
# AREA MEAN - LMDZ6A
# AUSTRAL
prwlmdz6a_austral = np.mean(prwlmdz6a_austral, axis=(1,2))
prlwlmdz6a_austral = np.mean(prlwlmdz6a_austral, axis=(1,2))
prswlmdz6a_austral = np.mean(prswlmdz6a_austral, axis=(1,2))
lmdz6a_austral = np.ma.dstack((prwlmdz6a_austral,prlwlmdz6a_austral,prswlmdz6a_austral))
# ANTARCTICA
prwlmdz6a_antarctica = np.mean(prwlmdz6a_antarctica, axis=(1,2))
prlwlmdz6a_antarctica = np.mean(prlwlmdz6a_antarctica, axis=(1,2))
prswlmdz6a_antarctica = np.mean(prswlmdz6a_antarctica, axis=(1,2))
lmdz6a_antarctica = np.ma.dstack((prwlmdz6a_antarctica,prlwlmdz6a_antarctica,prswlmdz6a_antarctica))
```

```python
months = ['Jan', 'Fev', 'Mar', 'Avr', 'Mai', 'Jun', 'Jui', 'Aou', 'Sep', 'Oct', 'Nov', 'Dec']
#months = np.linspace(1, 12, 12, dtype=int)
```

```python
fig = plt.figure(figsize=(16,9))
ax1 = plt.subplot2grid(shape=(2,4), loc=(0,0), colspan=2)
ax2 = plt.subplot2grid((2,4), (0,2), colspan=2)
ax3 = plt.subplot2grid((2,4), (1,1), colspan=2)
axes = [ax1, ax2, ax3]
pltcounter = 0
for plotid in axes :
    pltcounter = pltcounter + 1
    if pltcounter == 1:
        plotid.set_title("a) Precipitable water", size=15)
    elif pltcounter == 2:
        plotid.set_title("b) Precipitable liquid water", size=15)
    elif pltcounter == 3:
        plotid.set_title("c) Precipitable solid water", size=15)
        plotid.set_ylabel('kg/m2')
        plotid.set_xlabel('Month')
        
    #LMDZ6A
    cllmdz6a_austral_plot = lmdz6a_austral[0,:,pltcounter-1]
    cllmdz6a_antarctica_plot = lmdz6a_antarctica[0,:,pltcounter-1]

    plotid.plot(months,cllmdz6a_austral_plot, c='b', linestyle='--', label='LMDZ6A austral ocean')
    ax_antarctica = plotid.twinx()
    ax_antarctica.plot(months,cllmdz6a_antarctica_plot, c='k', linestyle='--', label='LMDZ6A antarctica')
    ax_antarctica.set_ylabel('kg/m2')
    plotid.set_xticks(months, months, rotation='vertical')
    plotid.tick_params(axis='y', colors='blue')
    plotid.yaxis.label.set_color('blue')
    plotid.set_ylabel('kg/m2')
    plotid.set_xlim('Jan', 'Dec')

season_boundaries = [2.6, 5.7, 8.8]
def add_season_lines(ax):
    for boundary in season_boundaries:
        ax.axvline(x=boundary, color='gray', linestyle=':')
for plotid in axes:
    add_season_lines(plotid)
    
handles, labels = plotid.get_legend_handles_labels()
fig.legend(handles, labels, loc=(0.75,0.75))
handles, labels = ax_antarctica.get_legend_handles_labels()
fig.legend(handles, labels, loc=(0.75,0.8))
fig.suptitle('Seasonal cycle of area-integrated GOCCP and LMDZ6A precipitable water between {}Â°S and {}Â°S'.format(-int(lat_austral[-1]),-int(lat_austral[0])), fontsize=20)
fig.tight_layout()
path_fig = '/home/jcharrel/Documents/stage_M2/figures/CALIPSO_seasonal/'
plt.savefig(path_fig+'prw_seasonal_integrated_obs_lmdz6a.png',dpi=300)
```
